```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/content/acceptance_data.txt',names=['exam1','exam2','decision'])

# step 1
print(data.describe())
data.plot(kind='box',subplots=True)
plt.show()
```

```python
# step 2,3
x = data.drop(columns=['decision']).round(2)
y = data['decision']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2 , random_state=42)

# the normalization step but we dont need it here
# xmin = x_train.min()
# xmax = x_train.max()
# x_train = (x_train-xmin)/(xmax-xmin)
# x_test = (x_test-xmin)/(xmax-xmin)
```

```python
# reshaping
x_train_df = x_train.copy()
y_train_df = y_train.copy()

x_train = x_train.to_numpy().reshape((-1,2))
x_test = x_test.to_numpy().reshape((-1,2))
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()
```

```python
#step 4
from sklearn.naive_bayes import GaussianNB  #if data is numeric values
from sklearn.naive_bayes import CategoricalNB #if data is categories
# if mixed build 2 models w hcombine elresult bta3thom
# how to from continous to discrete
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
# step 5
print("Accuracy of NB: ",model.score(x_test,y_test))
```
